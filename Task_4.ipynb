```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data
data = {
    'Age': [25, 30, 35, 40, 45, 50, 55, 60, 65, 70],
    'Income': [30000, 35000, 40000, 45000, 50000, 55000, 60000, 65000, 70000, 75000],
    'Expenses': [15000, 18000, 20000, 22000, 24000, 26000, 28000, 30000, 32000, 34000],
}

df = pd.DataFrame(data)
```

```python
# Histogram for Age
plt.figure(figsize=(6, 4))
sns.histplot(df['Age'], kde=True)
plt.title("Distribution of Age")
plt.show()
```

```python
# Box plot for Income
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['Income'])
plt.title("Box Plot of Income")
plt.show()
```

```python
# Scatter plot between Age and Income
plt.figure(figsize=(6, 4))
sns.scatterplot(x='Age', y='Income', data=df)
plt.title("Age vs. Income")
plt.show()
```

```python
# Correlation matrix
correlation_matrix = df.corr()

# Heatmap to visualize correlations
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.violinplot(x='Age', y='Income', data=df)
plt.title("Violin Plot: Age vs. Income")
plt.show()
```
