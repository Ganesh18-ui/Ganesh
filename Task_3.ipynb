```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame with monthly sales data
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Sales": [100, 120, 150, 170, 190, 200],
}

df = pd.DataFrame(data)

# Compute growth rate as the percentage increase in sales from the previous month
df["Growth Rate (%)"] = df["Sales"].pct_change() * 100

# Create a figure with two subplots (bar chart and line chart)
fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar chart for sales data
ax1.bar(df["Month"], df["Sales"], color='b', label='Monthly Sales')
ax1.set_xlabel("Month")
ax1.set_ylabel("Sales", color='b')
ax1.set_title("Monthly Sales and Growth Rate")
ax1.legend(loc="upper left")

# Create a secondary axis for the growth rate
ax2 = ax1.twinx()  # Create a twin Axes sharing the same x-axis
ax2.plot(df["Month"], df["Growth Rate (%)"], color='r', linestyle='-', marker='o', label='Growth Rate')
ax2.set_ylabel("Growth Rate (%)", color='r')
ax2.legend(loc="upper right")

# Display the plot
plt.show()
```
